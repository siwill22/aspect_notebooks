```python
import matplotlib.pyplot as plt
import numpy as np

BASE_FILENAME = 'MOR_simple_base.prm'
NEW_FILENAME_STEM = 'MOR_simple'
BOX_CENTRE_X = 2500e3
BOX_CENTRE_Y = 2000e3

def rotate_via_numpy(xy, radians):
    """Use numpy to build a rotation matrix and take the dot product."""
    x, y = xy
    c, s = np.cos(radians), np.sin(radians)
    j = np.matrix([[c, s], [-s, c]])
    m = np.dot(j, [x, y])

    return float(m.T[0]), float(m.T[1])

def create_prm_file(transform_offset, transform_angle, plotting=False):
    
    XY = [(-3000e3,0+transform_offset/2.), (0,0+transform_offset/2.), 
          (0,0-transform_offset/2.), (3000e3,0-transform_offset/2.)]

    XYrot = []
    for xy in XY:
        XYrot.append(rotate_via_numpy(xy, np.radians(transform_angle)))
    
    if plotting:
        plt.plot(zip(*XYrot)[0],zip(*XYrot)[1])
        plt.gca().set_aspect('equal')
        plt.show()
    
    f = open(BASE_FILENAME)
    nf = open('%s_offset%dkm_angle%d.prm' % (NEW_FILENAME_STEM,transform_offset/1000.,transform_angle),'w')
    for line in f:
        if 'set Ridge line segments =' in line:
            ridge_coordinate_string = '%0.2f,%0.2f>%0.2f,%0.2f; %0.2f,%0.2f>%0.2f,%0.2f' % (XYrot[0][0]+BOX_CENTRE_X/2.,
                                                                                            XYrot[0][1]+BOX_CENTRE_Y/2.,
                                                                                            XYrot[1][0]+BOX_CENTRE_X/2.,
                                                                                            XYrot[1][1]+BOX_CENTRE_Y/2.,
                                                                                            XYrot[2][0]+BOX_CENTRE_X/2.,
                                                                                            XYrot[2][1]+BOX_CENTRE_Y/2.,
                                                                                            XYrot[2][0]+BOX_CENTRE_X/2.,
                                                                                            XYrot[2][1]+BOX_CENTRE_Y/2.)
            newline = line.replace(line, 'set Ridge line segments = %s' % ridge_coordinate_string)
            nf.write(newline)
        else:
            nf.write(line)
        
```

```python
transform_offset = 500e3

for transform_angle in np.arange(-45,46,15):

    create_prm_file(transform_offset, transform_angle, plotting=True)

```
